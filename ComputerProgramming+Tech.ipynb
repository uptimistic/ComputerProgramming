# The Language of a Computer-
 Machine Language, ASCII, UTF-8, UNICODE,
 Memory representation of data (encoding)





*   Computer is an electronic device. 
*   Electrical signals are used inside the computer to process information.
*   Digital signals represent information with a sequence of 0s and 1s.
*   The sequence of 0s and 1s called machine language.
*   Machine language is also referred to as low-level language
*   Human-readable data or information is high-level language
*   Transition from high-level to low-level language is done by compiler or intepretor

*   Compiler converts/translates the whole instruction set in a code into machine code before run 
> Takes time to get everything translated but runs fast 

> Errors are caught at run time 

*   Interpreter converts/translates each line of the whole instruction set in a code into machine code one at a time before run 
> Takes time to get everything translated but runs fast 
> Errors are caught at each time instruction line of code is written
